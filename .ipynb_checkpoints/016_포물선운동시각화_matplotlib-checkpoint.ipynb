## 만유인력에 대한 뉴턴의 법칙 시각화
$$F = G \frac{m_1 m_2}{r^2}$$
여기서 F는 두 물체 사이의 중력, G는 중력 상수, m1, m2는 두 물체의 질량, 그리고 r은 두 물체의 중심 사이의 거리를 나타냅니다. 이 수식은 두 물체 간의 중력이 질량에 비례하고 거리의 제곱에 반비례한다는 것을 보여줍니다.

```python
import matplotlib.pyplot as plt
```

```python
# 중력상수 G
G = 6.674*(10**-11)

# 두개의 질량값
m1 = 0.5
m2 = 1.5

# 거리에 대한 List 생성
r = range(100, 1001, 50)

# 계산된 F 값을 저장하기 위한 List
F = []

# 거리별 F를 계산하고 list에 저장
for dist in r:
    force = (G * m1 * m2 / dist**2)
    F.append(force)

plt.plot(r, F, marker='o')
plt.xlabel('Distance')
plt.ylabel('Force')
plt.title('Distance vs. Force')
plt.show()
```

## 포물선 운동 시각화

아래와 같이 $v_0$의 속도로 $\theta$의 각으로 물체가 투사된 경우, 모든 마찰을 무시한다면, 물체는 포물선 궤도로 나아가게 된다.
A,B는 각각 궤도의 최고점, O에서 투사된 후 다시 $x$축에 도달한 점이다. H는 점 A에서 $x$축에 내린 수선의 발이다.

포물선 운동의 방정식 ([유도 과정](https://namu.wiki/w/%ED%8F%AC%EB%AC%BC%EC%84%A0%20%EC%9A%B4%EB%8F%99?rev=90))
- 포물선 운동 시간(체공 시간): $T = \frac{2v_0}{g} \sin\theta$
- 수평 도달 거리: $R = \frac{v_0^2}{g} \sin 2\theta$
- 최고점 높이: $H = \frac{v_0^2 \sin^2 \theta}{2g}$
- 시간 경과에 따른 물체의 위치:   
$$x(t) = v_0 t \cos\theta$$
$$y(t) = v_0 t \sin\theta - \frac{1}{2} g t^2$$
이 두식에서 t를 소거하면 궤도 방정식은 다음과 같은 포물선 방정식이 된다.
$$y(x) = x\tan\theta - \frac{g}{2v_0^2\cos^2\theta}x^2$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# 중력 가속도 (9.81m/s^2)
g = 9.81
```

```python
"""
공중에 투척한 물체의 포물서 궤적 그리기
"""
try:
    # 발사 속도 (m/s)
    v = float(input("초기 발사 속도(m/s)를 입력하세요: "))

    # 발사각 (degree)
    degree = float(input("발사각(degree)을 입력하세요: "))
except ValueError:
    print("숫자를 입력하세요.")

print()
print(f"발사 속도 {v}m/s, 발사각{degree}도")
```

```python
plt.figure(figsize=(12, 4))

def plot_trajectory(v_0, angle):
    #발사각(radian)
    theta = math.radians(angle)

    #포물선 운동 시간(체공 시간) 계산하기
    t_flight = 2*v_0 * math.sin(theta) / g

    #시간 간격 계산하기
    intervals = np.arange(0, t_flight, 0.001)

    # x, y 좌표 list
    x, y = [], []
    for t in intervals:
        x.append(v_0 * t * math.cos(theta))
        y.append(v_0 * t * math.sin(theta) - 0.5 * g * t**2)

    plt.plot(x, y)
    plt.xlabel('distance')
    plt.ylabel('height')
    plt.title('Projection motion of an object')

plot_trajectory(v, degree)
```

### 서로 다른 초기 발사 속도에 따른 궤적 비교

```python
v_list = [20, 40, 60]
degree = 45

plt.figure(figsize=(12, 4))

for v in v_list:
    plot_trajectory(v, degree)
    plt.legend(v_list)
```

### 서로 다른 발사각에 따른 궤적 비교

```python
angles = [15, 30, 60]

v_0 = 25

plt.figure(figsize=(12, 4))

for angle in angles:
    plot_trajectory(v_0, angle)
    plt.legend(angles)
```
