### 함수를 시각화

`plot()`는 SymPy를 사용하여 수학적 표현식의 그래프를 그릴 때 사용됩니다. 여기서 `plot` 함수에 전달된 파라미터들에 대한 설명은 다음과 같습니다:

- 첫 번째 인자는 그래프로 그릴 단일 변수 함수를 나타내는 표현식입니다.

- 두 번째 인자는 자유 변수의 범위를 나타내는 3-튜플입니다. 예를 들어, `(x, 0, 5)`는 변수 `x`의 범위가 0부터 5까지임을 의미합니다.

- `show=False`: 이 옵션은 그래프를 즉시 보여주지 않고, 대신 그래프 객체를 반환하여 나중에 수정하거나 조회할 수 있도록 합니다. `True`로 설정하면 함수가 호출될 때 그래프가 즉시 표시됩니다. `False`로 설정하면, `p.show()`와 같은 명령을 사용하여 명시적으로 그래프를 표시해야 합니다.

```python
# x를 수학적 심볼로 정의
x = Symbol("x")
y = x**2

# x에 대한 제한이 없으므로 기본 범위에서 y=x**2의 그래프를 그림
p = plot(y)
print(p)
```

```python
# x**2의 그래프를 그리고, x의 범위를 -3에서 3으로 설정
p1 = plot(x**2, (x, -3, 3))
```

p는 plot 함수에 의해 반환된 그래프 객체이며, p[0]는 이 객체 내에서 첫 번째 그래프 선을 가리킵니다.

```python
# 그래프의 첫 번째 선의 색상을 빨간색으로 설정
p1[0].line_color = 'r'
p1.show()
```

```python
p = plot(y, show=False)
p.xlim = [0, 3]
p.ylim = [0, 2]
p[0].line_color = 'y'
p.show()
```

```python
x, a = symbols('x a')
expr = a/x
expr
```

```python
# 심볼 정의
x, a = symbols('x a')

# a를 포함하는 식을 가정합니다.
expr = a*x**2 + 2*x - 5

# 표현식에서 a=1로 대체하여, x의 범위가 -5부터 5까지인 플롯 생성
p = plot(expr.subs(a, 1), (x, -5, 5), show=False)

# 플롯의 레이블 설정: 플롯에 식을 표시
p[0].label = f'$y={latex(expr.subs(a, 1))}$'

# y축의 범위 설정
p.ylim = [-7, 5]

p.legend = True
p.show()
```

### matplotlib + sympy expression

```python
# 심볼 x 정의
x = symbols('x')

# 함수 y 정의
y = x**2 - 3*x

# -11부터 11까지의 x 값에 대해 y 값을 계산하고, 각 점을 그래프에 표시
for pt in range(-11, 12):
    plt.plot(pt, y.subs(x, pt), 'o')  # 'o'는 점으로 표시하라는 의미

plt.xlabel('x')
plt.ylabel('y')

# 그래프 제목 설정: 함수 형태를 제목으로 표시
plt.title(f'f(x)={y}')

plt.show()
```

 ### 주어진 이차방정식 $ax^2 + bx + c$의 근을 찾고, 해당 이차방정식의 그래프를 그리는 함수를 정의

 판별식(Discriminant)은 이차방정식의 뿌리(근)의 성질을 판별하는 데 사용되는 수학적 표현입니다. 이차방정식이 $ax^2 + bx + c=0$ 형태일 때, 해당 판별식은 $D=b^2−4ac$로 정의됩니다.

판별식의 값에 따라 이차방정식의 근의 성질이 결정됩니다:

판별식 > 0: 이 경우, 이차방정식은 서로 다른 두 실수 근을 가집니다.  
판별식 = 0: 이 경우, 이차방정식은 중근을 가집니다 (서로 같은 두 실수 근).  
판별식 < 0: 이 경우, 이차방정식은 실수 근이 없으며, 두 복소수 근을 가집니다.

근의 공식은 $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ 입니다.

```python
import numpy as np

# 이차방정식의 근을 계산하는 함수
def calc_quadratic_roots(a, b, c):
    D = b**2 - 4*a*c              # 판별식 계산
    x1 = (-b + np.sqrt(D)) / (2*a)       # 첫 번째 근 계산
    x2 = (-b - np.sqrt(D)) / (2*a)        # 두 번째 근 계산

    print("첫번째 root = ", x1)
    print("두번째 root =", x2)

# 이차방정식의 그래프를 그리는 함수
def printGraph(a, b, c):
    x = Symbol("x")
    plot(a*x**2 + b*x + c)
```

```python
# 매개변수 설정
a = 1
b = 2
c = 1

# 함수 호출
calc_quadratic_roots(a, b, c)  # 이차방정식의 근 계산
printGraph(a, b, c)  # 이차방정식의 그래프 그리기
```

```python
a = 1
b = 0
c = -6
calc_quadratic_roots(a, b, c)
printGraph(a, b, c)
```

### 로그 표현
$\log_{b}{x}$는 SymPy 함수 log(x, b)로 얻을 수 있습니다.  
이 함수의 첫 번째 매개변수는 인수이고, 두 번째 매개변수는 밑(base)입니다.  

log(x, b)는 $\log_{b}{x}$ 대신 $\frac{log(x)}{log(b)}$를 반환합니다.   

Python 에서 log 는 자연로그를 의미합니다.

```python
# 심볼 b, x 정의
b, x = symbols('b x')

# 밑이 b이고, 변수가 x인 로그함수 정의
log_func = log(x, b)
log_func
```

```python
b = 10
plot(log(x, b))
```

두 번째 매개변수를 생략하고 단순히 log(x)를 쓰면 자연 로그가 됩니다.

함수 ln(x)는 자연 로그에도 사용할 수 있습니다.

```python
ln(x)
```

```python
plot(ln(x))
```

```python
# x 값을 0.001부터 16까지 2000개의 점으로 생성
x = np.linspace(0.001, 16, 2000)

y_e = np.log(x)   # 자연로그 ln(x)
y_2 = np.log2(x)  # 밑이 2인 로그 log2(x)
y_10 = np.log10(x)  # 밑이 10인 로그 log10(x)

plt.plot(x, y_e)  # 자연로그 그래프
plt.plot(x, y_2)  # 밑이 2인 로그 그래프
plt.plot(x, y_10)  # 밑이 10인 로그 그래프

# 범례 추가
plt.legend(['ln', 'log2', 'log10'])

# x축과 y축에 선 추가 (원점을 기준으로)
plt.axhline(0, color='black', linewidth=0.5)  # y=0에 수평선 추가
plt.axvline(0, color='black', linewidth=0.5)  # x=0에 수직선 추가

# 그래프 표시
plt.show()
```

### 문제 2)
두 점의 좌표 (a,b)와 (c,d)가 주어졌을 때, 이 두 점을 지나는 직선의 방정식을 반환하는 함수 linear(a, b, c, d)를 정의하라

- 기울기 : $slope = \frac{(d - b)}{(c - a)}$

- "점-기울기 형식(Point-Slope Form)"은 직선의 방정식을 표현하는 한 방법으로, 한 점과 그 직선의 기울기를 사용하여 직선의 방정식을 나타냅니다. 이 형식은 다음과 같은 수학적 공식으로 표현됩니다:  
$$y−y_1=m(x−x_1)$$

여기서 $(x_1, y_1)$은 직선 위의 한 점, $m$은 직선의 기울기를 나타냅니다. 이 공식을 사용하면, 직선 위의 어떤 점과 그 점을 지나는 직선의 기울기가 주어졌을 때, 그 직선의 방정식을 쉽게 구할 수 있습니다.  

예를 들어, 직선이 점 $(a,b)$를 지나고 기울기가 $m$일 때, 이 직선의 방정식은 $y−b=m(x−a)$로 표현됩니다. 이 방정식은 직선의 모든 점 $(x,y)$에 대해 성립합니다.
