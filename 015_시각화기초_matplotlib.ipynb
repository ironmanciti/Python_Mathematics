패치를 사용하면 점과 선 이상의 그래픽 요소를 시각화에 추가할 수 있어 데이터 시각화의 표현력을 크게 향상시킬 수 있습니다.

- patches.Circle : 원을 생성  
- add_patch : 생성한 도형(사각형, 원 등)을 그래프의 축(ax)에 추가합니다. 이를 통해 도형이 시각적으로 표시됩니다.  
- patches.Rectangle : 사각형을 생성  

```python
import matplotlib.patches as patches

"""
사각형 내에 원 그리기 함수
"""
def draw_circle(ax, x, y):
    # 주어진 좌표 (x, y)에 원을 그리는 함수
    circle = patches.Circle((x, y), 0.5, edgecolor='black', facecolor='blue')
    ax.add_patch(circle) # 축에 원 추가

# 그래프 및 사각형 설정
fig, ax = plt.subplots(figsize=(5, 5))  # 5x5 크기의 그래프 생성
ax.set_xlim(0, 6)  # x축 범위 설정
ax.set_ylim(0, 6)  # y축 범위 설정

# 사각형의 좌측 하단 모서리의 위치 (x, y), 너비, 높이, 테두리 색상, 내부 색상
square = patches.Rectangle((1, 1), 4, 4, edgecolor='black', facecolor='none')
ax.add_patch(square)  # 축에 사각형 추가

# 사각형 내에 원 그리기
# 각 원의 반지름이 0.5이므로, 원이 사각형의 경계선에 닿지 않기 위해 y = 1.5에서 시작
y = 1.5
while y < 5:
    x = 1.5
    while x < 5:
        draw_circle(ax, x, y)  # 원 그리기 함수 호출
        x += 1.0  # x 좌표 증가
    y += 1.0  # y 좌표 증가

plt.show()
```
